```python
#7
list1 = [1,2,2,3,33,4,4,5,4,3,2,1]

def countX(list1, x):
    count = 0
    for ele in list1:
        if (ele == x):
            count = count + 1
    return count
x = 4
print('{} has occurred {} times '.format(x,countX(list1,x)))
```
